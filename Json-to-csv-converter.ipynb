```python
import json
import pandas as pd

# Load your JSON file
with open("HDFC Bank 15 min.json", "r") as f:
    data = json.load(f)

# Convert JSON arrays to DataFrame
df = pd.DataFrame({
    "timestamp": data["t"],
    "open": data["o"],
    "high": data["h"],
    "low": data["l"],
    "close": data["c"],
    "volume": data["v"]
})

# Convert UNIX timestamp to readable datetime
df["datetime"] = pd.to_datetime(df["timestamp"], unit="s")

# Arrange columns properly
df = df[["datetime", "timestamp", "open", "high", "low", "close", "volume"]]

# Save to CSV
df.to_csv("HDFC_Bank_15min.csv", index=False)

print("CSV file saved successfully!")
```
